```python
df['Support Devices'].value_counts()
```

```python
# Important Note: Labels are BELOW the data
for col in df.columns:
   print( df[col].value_counts())
```

```python
df['Edema'].value_counts().plot(kind='barh')
```

```python
pieces = []
for col in df.columns:
    tmp_series = df[col].value_counts()
    tmp_series.name = col
    pieces.append(tmp_series)
df_value_counts = pd.concat(pieces, axis=1)
```
